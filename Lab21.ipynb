```python
from bs4 import BeautifulSoup
import requests ## HTTP REQUEST
import pandas as pd
```

```python
url='https://www.billboard.com/charts/hot-100/'
```

```python
headers={'User-Agent':'MiAplication/1.0'}
```

```python
response=requests.get(url, headers=headers)
```

```python
response.status_code
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
```

```python
type(soup)
```

```python
print(soup.get_text())
```

```python
soup.find_all('h3', class_='a-font-primary-bold-s', id="title-of-a-story")
```

```python
songs3= soup.find_all('h3', class_='a-font-primary-bold-s', id="title-of-a-story")
```

```python
song_titles = []

for i in songs3:
    song = i.get_text()
    song_titles.append(song)
```

```python
song_titles
```

```python
len(song_titles)
```

```python
clean_song_titles = [song.replace('\n','').replace('\t','') for song in song_titles]
```

```python
clean_song_titles[2:102]
```

```python
songs = clean_song_titles[2:102]
```

```python
soup.find_all('span', class_='a-no-trucate')
```

```python
artists = soup.find_all('span', class_='a-no-trucate')
```

```python
artists_list = []

for i in artists:
    artist = i.get_text()
    artists_list.append(artist)
```

```python
artists_list
```

```python
len(artists_list)
```

```python
clean_artists = [artist.replace('\n','').replace('\t','') for artist in artists_list]
```

```python
clean_artists
```

```python
billboard=pd.DataFrame({'artist':clean_artists,
                   'song':songs})
```

```python
billboard
```

```python
def search_song (billboard):
    song_query = input('Type the title of the song you are looking for: ').lower()
   
    if len(song_query) < 3:
       print('Entry should have at least 3 characters')
       return

    results = billboard[billboard['song'].str.lower().str.contains(song_query)]

    if len(results) > 0:
        print('The following songs match with your query: ')
        print(results)

        confirmation = input('Is this one of the songs you were looking for? (Yes/No): ').lower()

        if confirmation in ['yes','si','y','sí']:
            for index, row in results.iterrows():
                print(f'Song: {row["song"]}')
                print(f'Artist: {row["artist"]}')
                print('Your song is on Billboards Top 100!')
        else:
            print('I was not able to find your song')
    else:
        print('I was not able to find your song')

                
```

```python
search_song (billboard)
```
